```python
import pandas as pd
import seaborn as sns                                                                                     
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
sns.set_palette("rocket_r")
```

```python
results = pd.read_csv("results.csv")
results.head()

results["per_iteration"] = results["time"] / 100000 * 1000
```

```python
sns.boxplot(x="algorithm", y="per_iteration", data=results, showfliers=False)

plt.xlabel(None)
plt.ylabel('Seconds per 1K Iterations')

plt.xticks(rotation=45)
```
