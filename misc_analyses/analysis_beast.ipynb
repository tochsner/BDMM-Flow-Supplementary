```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_theme(style="darkgrid")
```

```python
results = pd.read_csv("results_beast.csv")
results.time = results.time / 100_000
```

```python
sns.barplot(x="algorithm", y="time", data=results, errorbar=None)
plt.xticks(rotation=45)
plt.xlabel('Algorithm')
plt.ylabel('Runtime s/iteration')
```
