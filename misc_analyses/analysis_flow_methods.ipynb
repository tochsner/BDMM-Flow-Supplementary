```python
import pandas as pd
import seaborn as sns                                                                                     
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_theme(style="darkgrid")
```

```python
normal = pd.read_csv("results_normal.csv")
random = pd.read_csv("results_normal_random.csv")
cached = pd.read_csv("results_cached.csv")
inverse = pd.read_csv("results_inverse.csv")
backwards = pd.read_csv("results_backwards.csv")
cached_backwards = pd.read_csv("results_cached_backwards.csv")
qr_cached_backwards = pd.read_csv("results_qr_cached_backwards.csv")
lu_cached_backwards = pd.read_csv("results_lu_cached_backwards.csv")
```

```python
ax = sns.barplot(data=pd.DataFrame({
    "Method": [
        "Normal", 
        "Random",
        "Random + Flow Cached",
        "QR Cached Backwards",
    ],
    "Rel. performance": [
        (normal["bdmm_duration"] / normal["flow_duration"]).mean(),
        (random["bdmm_duration"] / random["flow_duration"]).mean(),
        (cached["bdmm_duration"] / cached["flow_duration"]).mean(),
        (qr_cached_backwards["bdmm_duration"] / qr_cached_backwards["flow_duration"]).mean(),
    ]
}), x="Method", y="Rel. performance")
plt.xticks(rotation=45)
for i in ax.containers:
    ax.bar_label(i,)
```

```python
ax = sns.barplot(data=pd.DataFrame({
    "Method": [
        "Normal",
        "Random Initial Matrix",
        "Random + Flow Cached",
        "Random + QR Cached + Inverse Flow",
    ],
    "Rel. performance (> 100 leaves)": [
        (normal[normal["leaf_count"] > 100]["bdmm_duration"] / normal[normal["leaf_count"] > 100]["flow_duration"]).mean(),
        (random[random["leaf_count"] > 100]["bdmm_duration"] / random[random["leaf_count"] > 100]["flow_duration"]).mean(),
        (cached[cached["leaf_count"] > 100]["bdmm_duration"] / cached[cached["leaf_count"] > 100]["flow_duration"]).mean(),
        (qr_cached_backwards[qr_cached_backwards["leaf_count"] > 100]["bdmm_duration"] / qr_cached_backwards[qr_cached_backwards["leaf_count"] > 100]["flow_duration"]).mean(),
    ]
}), x="Method", y="Rel. performance (> 100 leaves)")
plt.xticks(rotation=45)

for i in ax.containers:
    ax.bar_label(i,)
```

```python
ax = sns.barplot(data=pd.DataFrame({
    "Method": [
        "Normal",
        "Random Initial Matrix",
        "Random + Flow Cached",
        "Random + QR Cached + Inverse Flow",
    ],
    "Rel. performance (> 500 leaves)": [
        (normal[normal["leaf_count"] > 500]["bdmm_duration"] / normal[normal["leaf_count"] > 500]["flow_duration"]).mean(),
        (random[random["leaf_count"] > 500]["bdmm_duration"] / random[random["leaf_count"] > 500]["flow_duration"]).mean(),
        (cached[cached["leaf_count"] > 500]["bdmm_duration"] / cached[cached["leaf_count"] > 500]["flow_duration"]).mean(),
        (qr_cached_backwards[qr_cached_backwards["leaf_count"] > 500]["bdmm_duration"] / qr_cached_backwards[qr_cached_backwards["leaf_count"] > 500]["flow_duration"]).mean(),
    ]
}), x="Method", y="Rel. performance (> 500 leaves)")
plt.xticks(rotation=45)

for i in ax.containers:
    ax.bar_label(i,)
```

```python
ax = sns.barplot(data=pd.DataFrame({
    "Method": [
        "Normal",
        "Random Initial Matrix",
        "Random + Flow Cached",
        "Random + QR Cached + Inverse Flow",
    ],
    "Rel. performance (> 100 leaves)": [
        ((normal["bdmm_likelihood"] - normal["flow_likelihood"]) / normal["bdmm_likelihood"]).abs().mean(),
        ((random["bdmm_likelihood"] - random["flow_likelihood"]) / random["bdmm_likelihood"]).abs().mean(),
        ((cached["bdmm_likelihood"] - cached["flow_likelihood"]) / cached["bdmm_likelihood"]).abs().mean(),
        ((qr_cached_backwards["bdmm_likelihood"] - qr_cached_backwards["flow_likelihood"]) / qr_cached_backwards["bdmm_likelihood"]).abs().mean(),
    ]
}), x="Method", y="Rel. performance (> 100 leaves)")
plt.xticks(rotation=45)

for i in ax.containers:
    ax.bar_label(i,)
```

```python
ax = sns.scatterplot(normal, x="flow_likelihood", y="bdmm_likelihood")
ax.plot(ax.get_xlim(), ax.get_xlim(), linestyle='--', color='k', lw=0.5, scalex=False, scaley=False)

plt.xlabel('Flow Log Likelihood')
plt.ylabel('BDMM Log Likelihood')

plt.yscale("log")
plt.xscale("log")

((cached["bdmm_likelihood"] - cached["flow_likelihood"]) / cached["bdmm_likelihood"]).abs().mean()
```
