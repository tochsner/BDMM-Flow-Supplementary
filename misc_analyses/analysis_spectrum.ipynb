```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

```python
no_random = pd.read_csv(
    "logs_spectrum_no_random.csv", 
    names=["run", "time", "trace", "norm", "det", "minEV", "maxEV", "minS", "maxS"]
)
random = pd.read_csv(
    "logs_spectrum_random.csv",
    names=["run", "time", "trace", "norm", "det", "minEV", "maxEV", "minS", "maxS"]
)
no_intervals = pd.read_csv(
    "logs_spectrum_no_random_no_intervals.csv",
    names=["run", "time", "trace", "norm", "det", "minEV", "maxEV", "minS", "maxS"]
)
```

```python
def add_step_sizes(df: pd.DataFrame):
    df.sort_values(["run", "time"], inplace=True)
    df["step_size"] = df["time"] - df["time"].shift(1)

add_step_sizes(no_random)
add_step_sizes(random)
add_step_sizes(no_intervals)
```

```python
sns.scatterplot(no_random.sample(10000), x="time", y="trace", alpha=0.1)
sns.scatterplot(no_intervals.sample(10000), x="time", y="trace", alpha=0.1)
sns.scatterplot(random.sample(10000), x="time", y="trace", alpha=0.1)
plt.legend(["I", "Random", "No Intervals or random"])
```

```python
sns.scatterplot(no_random.sample(10000), x="time", y="det", alpha=0.1)
sns.scatterplot(random.sample(10000), x="time", y="det", alpha=0.1)
sns.scatterplot(no_intervals.sample(10000), x="time", y="det", alpha=0.1)
plt.yscale("log")
plt.legend(["I", "Random"])
```

```python
sns.scatterplot(no_random.sample(10000), x="time", y="det", alpha=0.1)
sns.scatterplot(random.sample(10000), x="time", y="det", alpha=0.1)
plt.legend(["I", "Random"])
```

```python
sns.scatterplot(no_random.sample(10000), x="time", y="norm", alpha=0.1)
sns.scatterplot(random.sample(10000), x="time", y="norm", alpha=0.1)
plt.legend(["I", "Random"])
```

```python
sns.scatterplot(no_random.sample(10000), x="time", y="minEV", alpha=0.1)
sns.scatterplot(random.sample(10000), x="time", y="minEV", alpha=0.1)
plt.legend(["I", "Random"])
plt.yscale("log")
```

```python
sns.scatterplot(no_random.sample(10000), x="time", y="maxEV", alpha=0.1)
sns.scatterplot(random.sample(10000), x="time", y="maxEV", alpha=0.1)
plt.legend(["I", "Random"])
plt.yscale("log")
```

```python
sns.scatterplot(no_intervals.sample(10000), x="time", y="maxS", alpha=0.1)
sns.scatterplot(random.sample(10000), x="time", y="maxS", alpha=0.1)
sns.scatterplot(no_random.sample(10000), x="time", y="maxS", alpha=0.1)
plt.legend(["Normal", "Random", "I"])
plt.yscale("log")
```

```python
sns.scatterplot(no_intervals.sample(10000), x="time", y="step_size", alpha=0.5)
plt.legend(["Normal", "Random", "I"])
plt.yscale("log")
plt.ylim(1e-11, 1e-1)
```

```python
sns.scatterplot(no_random.sample(10000), x="time", y="step_size", alpha=0.05)
sns.scatterplot(random.sample(10000), x="time", y="step_size", alpha=0.05)
plt.legend(["I", "Random"])
plt.yscale("log")                 
plt.ylim(1e-11, 1e-1)
```
